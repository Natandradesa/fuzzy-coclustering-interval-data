```python
import pandas as pd
from pathlib import Path
```

```python

def display_average(averages, stds, decimals=2):
    """
    Combine averages and standard deviations into the format:
    mean (std).

    Parameters
    ----------
    averages : pandas.DataFrame
        Mean values.

    stds : pandas.DataFrame
        Standard deviation values.

    decimals : int, default=3
        Number of decimal places.

    Returns
    -------
    pandas.DataFrame
        Formatted table.
    """
    result = averages.copy().astype(str)

    for col in averages.columns:
        result[col] = [
            f"{avg:.{decimals}f} ({sd:.{decimals}f})"
            for avg, sd in zip(
                averages[col],
                stds[col],
            )
        ]

    return result



def read_simulation(path, methods, metrics):
    """
    Read simulation results and summarize them as mean (std).

    Parameters
    ----------
    path : str
        Directory containing the result files.

    methods : list of str
        Method names.

    metrics : list of str
        Metrics to summarize.

    Returns
    -------
    pandas.DataFrame
        Table with entries formatted as mean (std).
    """
    path = Path(path)

    averages = pd.DataFrame(index=methods, columns=metrics, dtype=float)
    stds = pd.DataFrame(index=methods, columns=metrics, dtype=float)

    for method in methods:

        file_path = path / f"{method}.txt"

        if not file_path.exists():
            raise FileNotFoundError(
                f"File not found:\n{file_path}"
            )

        results = pd.read_csv(file_path)
        results = results.loc[:, metrics]
        averages.loc[method] = results.mean()
        stds.loc[method] = results.std()
    return display_average(averages=averages, stds=stds)

```

### Synthetic Dataset 1

```python
path = '../synthetic_application/results/synthetic_1'
methods= ['IFCM', 'KFCM-IV', 'IFDK', 'IKFDK-O', 'IKFDK-K']
metrics = ['ARI', 'ARI_col', 'CARI']
X =read_simulation(path=path, methods=methods, metrics=metrics)
X
```
